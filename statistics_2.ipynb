**1. What is hypothesis testing in statistics?**

Hypothesis testing is a method used to decide whether there's enough evidence to support a specific claim about a population. It involves formulating a null and an alternative hypothesis, then using sample data to accept or reject the null.


**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

The null hypothesis (H₀) states there is no significant effect or relationship between variables, representing the default or "no difference" scenario, while the alternative hypothesis (H₁ or Hₐ) proposes there is a significant effect or relationship, which the researcher aims to prove by rejecting the null

**3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**


The significance level (alpha, α) in hypothesis testing is the researcher-set probability of rejecting a true null hypothesis, also known as a Type I error. It serves as the decision threshold: if the p-value from the test is less than α, the null hypothesis is rejected, indicating a statistically significant result; otherwise, it is not rejected. A lower α requires stronger evidence to reject the null hypothesis, while a higher α allows for less evidence but increases the risk of a Type I error.

Role in Deciding the Outcome

Setting the Threshold: Before any data is collected, the researcher chooses a significance level (e.g., 0.05 or 5%). This choice is made based on the specific field of study and the potential costs of a false positive.

Comparing p-value to α: After the experiment and statistical test, a p-value is calculated.

Making the Decision:
If the p-value < α: The result is considered statistically significant. The sample evidence is strong enough to reject the null hypothesis.

If the p-value ≥ α: The result is not considered statistically significant. The null hypothesis cannot be rejected, meaning there isn't enough evidence to conclude there is a real effect.

**4: What are Type I and Type II errors? Give examples of each.**

Type I Error (False Positive)
Definition: Rejecting the null hypothesis when it is actually true.

Example:

Hypothesis: A new marketing campaign has no effect on sales.

Error: A marketing manager concludes the new campaign significantly increased sales, when in reality, it had no impact.

Type II Error (False Negative)
Definition: Failing to reject the null hypothesis when it is actually false.

Example:
Hypothesis: A new medical test correctly identifies a disease.

Error: A patient receives a negative test result (test says "not sick") when they are, in fact, ill and the test should have identified the disease

**5: What is the difference between a Z-test and a T-test? Explain when to use each.**

A Z-test is used when the population standard deviation is known and the sample size is large (generally n ≥ 30), while a T-test is used when the population standard deviation is unknown and the sample size is small (generally n < 30)


You should use a Z-test when the following conditions are met:

Population Standard Deviation is Known: You have information about the true standard deviation of the entire population from which your sample was drawn.

Large Sample Size: Your sample size is typically 30 or more.
Normal Distribution: The population from which the sample is drawn is normally distributed.


You should use a T-test when:

Population Standard Deviation is Unknown: The standard deviation of the population is unknown, and you are using the sample standard deviation as an estimate.

Small Sample Size: Your sample size is small, generally fewer than 30.
Normal Distribution: The data is approximately normally distributed. The T-test accounts for the additional uncertainty that comes from not knowing the population standard deviation.

**Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.**

**(Include your Python code and output in the code box below.)**

 **Hint: Generate random number using random function.**


```python
import numpy as np
import matplotlib.pyplot as plt


n = 10  # Number of trials
p = 0.5 # Probability of success

num_samples = 10000
binomial_samples = np.random.binomial(n, p, num_samples)

# Plot the histogram
plt.hist(binomial_samples, bins=np.arange(-0.5, n + 1.5, 1),
         edgecolor='black', alpha=0.7, density=True)
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.xticks(np.arange(0, n + 1, 1)) # Set x-axis ticks to represent possible outcomes
plt.grid(axis='y', alpha=0.75)
plt.show()
```

7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50
sigma = 1
n = len(sample_data)

# Sample statistics
sample_mean = np.mean(sample_data)

# Z statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", round(sample_mean, 3))
print("Z-statistic:", round(z, 3))
print("p-value:", round(p_value, 5))

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis: Sample mean is significantly different from 50.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from 50.")
```

**8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=200)  # mean=50, std=5, sample size=200

# Step 2: Calculate statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# Step 3: Calculate 95% Confidence Interval
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # t critical value
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(sample_mean, color="red", linestyle="dashed", linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color="green", linestyle="dashed", linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color="green", linestyle="dashed", linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**


Explanation of Z-scores

A Z-score tells us how many standard deviations a data point is from the mean.

Formula:

𝑍 = 𝑥 − 𝜇 𝜎
Z= σ x−μ
	​
  
Interpretation:

Z = 0 → exactly at the mean.

Z = +1 → 1 standard deviation above the mean.

Z = -1 → 1 standard deviation below the mean.

Large |Z| (> 2 or 3) indicates outliers.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores and plot
def z_score_analysis(data):
    # Calculate mean and std deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation

    # Z-score formula
    z_scores = (data - mean) / std

    # Print first few Z-scores
    print("First 10 Z-scores:", np.round(z_scores[:10], 2))

    # Plot histogram of Z-scores
    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=15, color="skyblue", edgecolor="black", alpha=0.7)
    plt.axvline(0, color="red", linestyle="dashed", linewidth=2, label="Mean (0)")
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    return z_scores

# Example usage
np.random.seed(42)
sample_data = np.random.normal(100, 15, 200)  # mean=100, std=15
z_scores = z_score_analysis(sample_data)
```
